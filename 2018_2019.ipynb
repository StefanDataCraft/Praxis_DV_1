```python
# Stefan version
drehung = np.linspace(0, 359, 360)
for grad in drehung:
    ax = plt.axes(projection='3d')
    ax.scatter(with_german_columns_2018['BIP pro Kopf'], with_german_columns_2018['Gesunde Lebenserwartung'], with_german_columns_2018["Sozialhilfe"], c=with_german_columns_2018['Glcklichkeitswert'])
    cbar = fig.colorbar(scatter)
    ax.set_xlabel('BIP pro Kopf')
    ax.set_ylabel('Lebenserwartung')
    ax.set_zlabel('Sozialhilfe')
    ax.set_title('Glcklichkeitswert nach BIP pro Kopf und Lebenserwartung.')
    ax.view_init(20, grad)
    plt.savefig(fname=fr'export/rotation/bild{int(grad):0>3d}.png', dpi=72)
    print(f'glueckwrfel bei {grad} Grad Rotation gespeichert!')
    plt.close()
print('Gluekproduktion erfolgreich abgeschlossen!')
```

```python
#Chatgpt version

# Sample data
x = with_german_columns_2018['BIP pro Kopf']
y = with_german_columns_2018['Gesunde Lebenserwartung']
z = with_german_columns_2018["Sozialhilfe"]

colors = with_german_columns_2018['Glcklichkeitswert']
#cmap = plt.get_cmap('tab10')
drehung = np.linspace(0, 359, 360)

# Create a 3D axes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Loop through different rotations
for grad in drehung:
    ax.clear()
    scatter = ax.scatter(x, y, z, c=colors, cmap="magma")

    # Add a color bar to show the mapping of values to colors
    cbar = fig.colorbar(scatter)
    ax.set_xlabel('BIP pro Kopf')
    ax.set_ylabel('Lebenserwartung')
    ax.set_zlabel('Sozialhilfe')
    ax.set_title('Glcklichkeitswert nach BIP pro Kopf, Lebenserwartung und Sozialhilfe.')
    #ax.view_init(20, grad)
    ax.view_init(20, grad % 360)

    # Save the plot with a color legend
    plt.savefig(fname=f'export/rotation/bild{int(grad):0>3d}.png', dpi=72)
    print(f'glueckwrfel bei {grad} Grad Rotation gespeichert!')

print('Glueckproduktion erfolgreich abgeschlossen!')
```

```python
# Unabhngige Variablen auswhlen
X = with_german_columns_2018_2019[['Rang','BIP pro Kopf','Sozialhilfe','Gesunde Lebenserwartung','Entscheidungsfreiheit','Grosszgigkeit','Wahrnehmungen von Korruption']]

# Abhngige Variable auswhlen
y = with_german_columns_2018_2019['Glcklichkeitswert']

# Regressionsmodell erstellen und anpassen
regression_model = LinearRegression()
regression_model.fit(X, y)

# Koeffizienten und Intercept abrufen
coefficients = regression_model.coef_
intercept = regression_model.intercept_

# Ergebnisse anzeigen
print("Koeffizienten:", coefficients)
print("Intercept:", intercept)
```

```python
lnder = with_german_columns_2018['Land']
glcklichkeitswerte = with_german_columns_2018['Glcklichkeitswert']
bip_pro_kopf = with_german_columns_2018['BIP pro Kopf']  # Beispiel fr BIP pro Kopf
lebenserwartung = with_german_columns_2018['Gesunde Lebenserwartung']  # Beispiel fr Lebenserwartung
grosszgigkeit = with_german_columns_2018['Grosszgigkeit']  # Beispiel fr Bevlkerung in Millionen

# Skalieren der Bevlkerungsdaten fr die Bubble-Gre
bubble_size = np.array(grosszgigkeit) * 150  # Skalierung fr eine geeignete Darstellung

# Bubblechart erstellen
plt.figure(figsize=(10, 6))
plt.scatter(bip_pro_kopf, lebenserwartung, s=bubble_size, c=glcklichkeitswerte, cmap='viridis', alpha=0.7)

# Achsenbeschriftungen
plt.xlabel('BIP pro Kopf')
plt.ylabel('Lebenserwartung')

# Farbskala hinzufgen (Glcklichkeitswerte)
colorbar = plt.colorbar()
colorbar.set_label('Glcklichkeitswert')

# Beschriftungen fr die Lnder hinzufgen
#for i, land in enumerate(lnder):
#    plt.annotate(land, (bip_pro_kopf[i], lebenserwartung[i]), fontsize=10, ha='center', va='bottom')

# Titel hinzufgen
plt.title('Glcklichkeitswerte, BIP pro Kopf und Lebenserwartung.  Grsse ist Grosszgigkeit.')

plt.grid(True)
plt.savefig(f'export/Bubble1.png', dpi=300)
plt.show()
```

```python
# Gruppieren der Daten nach Jahr
grouped_data = with_german_columns_2018_2019.groupby('Jahr')

# Bubblechart erstellen
plt.figure(figsize=(12, 6))

for year, group in grouped_data:
    plt.scatter(
        x=group['BIP pro Kopf'],
        y=group['Gesunde Lebenserwartung'],
        s=group['Glcklichkeitswert'] * 50,  # Skalieren der Blasen fr den Glcklichkeitswert
        alpha=0.7,
        label=f'Jahr {year}'
    )

# Achsenbeschriftungen
plt.xlabel('BIP pro Kopf')
plt.ylabel('Lebenserwartung')

# Legende hinzufgen
plt.legend()

# Titel hinzufgen
plt.title('BIP pro Kopf und Lebenserwartungen nach Jahren. Grsse ist Glcklichkeitswert.')

plt.grid(True)
plt.savefig(f'export/Bubble2.png', dpi=300)
plt.show()
```
