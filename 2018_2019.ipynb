```python
# Stefan version
drehung = np.linspace(0, 359, 360)
for grad in drehung:
    ax = plt.axes(projection='3d')
    ax.scatter(with_german_columns_2018['BIP pro Kopf'], with_german_columns_2018['Gesunde Lebenserwartung'], with_german_columns_2018["Sozialhilfe"], c=with_german_columns_2018['Glücklichkeitswert'])
    cbar = fig.colorbar(scatter)
    ax.set_xlabel('BIP pro Kopf')
    ax.set_ylabel('Lebenserwartung')
    ax.set_zlabel('Sozialhilfe')
    ax.set_title('Glücklichkeitswert nach BIP pro Kopf und Lebenserwartung.')
    ax.view_init(20, grad)
    plt.savefig(fname=fr'export/rotation/bild{int(grad):0>3d}.png', dpi=72)
    print(f'glueckwürfel bei {grad} Grad Rotation gespeichert!')
    plt.close()
print('Gluekproduktion erfolgreich abgeschlossen!')
```

```python
#Chatgpt version

# Sample data
x = with_german_columns_2018['BIP pro Kopf']
y = with_german_columns_2018['Gesunde Lebenserwartung']
z = with_german_columns_2018["Sozialhilfe"]

colors = with_german_columns_2018['Glücklichkeitswert']
#cmap = plt.get_cmap('tab10')
drehung = np.linspace(0, 359, 360)

# Create a 3D axes
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Loop through different rotations
for grad in drehung:
    ax.clear()
    scatter = ax.scatter(x, y, z, c=colors, cmap="magma")

    # Add a color bar to show the mapping of values to colors
    cbar = fig.colorbar(scatter)
    ax.set_xlabel('BIP pro Kopf')
    ax.set_ylabel('Lebenserwartung')
    ax.set_zlabel('Sozialhilfe')
    ax.set_title('Glücklichkeitswert nach BIP pro Kopf, Lebenserwartung und Sozialhilfe.')
    #ax.view_init(20, grad)
    ax.view_init(20, grad % 360)

    # Save the plot with a color legend
    plt.savefig(fname=f'export/rotation/bild{int(grad):0>3d}.png', dpi=72)
    print(f'glueckwürfel bei {grad} Grad Rotation gespeichert!')

print('Glueckproduktion erfolgreich abgeschlossen!')
```

```python
# Unabhängige Variablen auswählen
X = with_german_columns_2018_2019[['Rang','BIP pro Kopf','Sozialhilfe','Gesunde Lebenserwartung','Entscheidungsfreiheit','Grosszügigkeit','Wahrnehmungen von Korruption']]

# Abhängige Variable auswählen
y = with_german_columns_2018_2019['Glücklichkeitswert']

# Regressionsmodell erstellen und anpassen
regression_model = LinearRegression()
regression_model.fit(X, y)

# Koeffizienten und Intercept abrufen
coefficients = regression_model.coef_
intercept = regression_model.intercept_

# Ergebnisse anzeigen
print("Koeffizienten:", coefficients)
print("Intercept:", intercept)
```

```python
länder = with_german_columns_2018['Land']
glücklichkeitswerte = with_german_columns_2018['Glücklichkeitswert']
bip_pro_kopf = with_german_columns_2018['BIP pro Kopf']  # Beispiel für BIP pro Kopf
lebenserwartung = with_german_columns_2018['Gesunde Lebenserwartung']  # Beispiel für Lebenserwartung
grosszügigkeit = with_german_columns_2018['Grosszügigkeit']  # Beispiel für Bevölkerung in Millionen

# Skalieren der Bevölkerungsdaten für die Bubble-Größe
bubble_size = np.array(grosszügigkeit) * 150  # Skalierung für eine geeignete Darstellung

# Bubblechart erstellen
plt.figure(figsize=(10, 6))
plt.scatter(bip_pro_kopf, lebenserwartung, s=bubble_size, c=glücklichkeitswerte, cmap='viridis', alpha=0.7)

# Achsenbeschriftungen
plt.xlabel('BIP pro Kopf')
plt.ylabel('Lebenserwartung')

# Farbskala hinzufügen (Glücklichkeitswerte)
colorbar = plt.colorbar()
colorbar.set_label('Glücklichkeitswert')

# Beschriftungen für die Länder hinzufügen
#for i, land in enumerate(länder):
#    plt.annotate(land, (bip_pro_kopf[i], lebenserwartung[i]), fontsize=10, ha='center', va='bottom')

# Titel hinzufügen
plt.title('Glücklichkeitswerte, BIP pro Kopf und Lebenserwartung.  Grösse ist Grosszügigkeit.')

plt.grid(True)
plt.savefig(f'export/Bubble1.png', dpi=300)
plt.show()
```

```python
# Gruppieren der Daten nach Jahr
grouped_data = with_german_columns_2018_2019.groupby('Jahr')

# Bubblechart erstellen
plt.figure(figsize=(12, 6))

for year, group in grouped_data:
    plt.scatter(
        x=group['BIP pro Kopf'],
        y=group['Gesunde Lebenserwartung'],
        s=group['Glücklichkeitswert'] * 50,  # Skalieren der Blasen für den Glücklichkeitswert
        alpha=0.7,
        label=f'Jahr {year}'
    )

# Achsenbeschriftungen
plt.xlabel('BIP pro Kopf')
plt.ylabel('Lebenserwartung')

# Legende hinzufügen
plt.legend()

# Titel hinzufügen
plt.title('BIP pro Kopf und Lebenserwartungen nach Jahren. Grösse ist Glücklichkeitswert.')

plt.grid(True)
plt.savefig(f'export/Bubble2.png', dpi=300)
plt.show()
```
